```python
import matplotlib.pyplot as plt
import numpy as np
#指数とか対数とか扱うのに使うライブラリ
import math
```

```python
#シグモイド関数
#linspaceでxの幅を自動で作ってくれる
x=np.linspace(-10,10)
#pow(x,y)でxのy乗
y=1/(pow(math.e,-x)+1)
plt.plot(x,y)
```

```python
#関数化
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
#arrangeは連番や等差数列を生成する関数
#linspaceは第三引数で要素数を指定し、arangeは第三引数で間隔を指定できる
x=np.arange(-6.0,6.0,0.1)
```

```python
y=sigmoid(x)
plt.plot(x,y)
plt.show()
```

```python
#ReLU関数
def relu(x):
    #makiximum(a,b)でaかbの大きい値を返す
    return np.maximum(0,x)
```

```python
x=np.arange(-5,5,0.1)
y=relu(x)
plt.plot(x,y)
plt.show()
```
